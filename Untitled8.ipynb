```python
import cv2
import re
import os
import numpy as np
from paddleocr import PaddleOCR
from ultralytics import YOLO

# ---------------- Configuration ----------------
YOLO_MODEL_PATH = "/content/runs/detect/train/weights/best.pt"  # Replace with your YOLOv8 model path
IMAGE_PATH = "/content/new56.jpg"  # Replace with your image path

# Regex pattern for Indian number plates
plate_pattern = re.compile(r'^[A-Z]{2}\d{1,2}[A-Z]{1,2}\d{1,4}$')

# ---------------- Model Initialization ----------------
model = YOLO(YOLO_MODEL_PATH)
ocr = PaddleOCR(use_angle_cls=True, lang='en', show_log=False)

# ---------------- Preprocessing Function ----------------
def preprocess_plate(plate_img):
    plate_img = cv2.resize(plate_img, (0, 0), fx=2.5, fy=2.5, interpolation=cv2.INTER_CUBIC)
    gray = cv2.cvtColor(plate_img, cv2.COLOR_BGR2GRAY)
    denoised = cv2.bilateralFilter(gray, 11, 17, 17)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    enhanced = clahe.apply(denoised)
    kernel_sharpening = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
    sharpened = cv2.filter2D(enhanced, -1, kernel_sharpening)
    _, thresh = cv2.threshold(sharpened, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
    morph = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
    padded = cv2.copyMakeBorder(morph, 15, 15, 15, 15, cv2.BORDER_CONSTANT, value=255)
    return padded

# ---------------- Annotation Function ----------------
def annotate_image(img, box, label, color=(0, 255, 0)):
    x1, y1, x2, y2 = box
    cv2.rectangle(img, (x1, y1), (x2, y2), color, 2)
    cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

# ---------------- Main Processing Pipeline ----------------
def process_image(image_path):
    image = cv2.imread(image_path)
    annotated_image = image.copy()
    number_plate_texts = []
    helmet_detected = False

    results = model(image_path)[0]

    for box in results.boxes:
        cls_id = int(box.cls[0])
        conf = float(box.conf[0])
        class_name = results.names[cls_id].lower()
        x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())

        if class_name == "helmet" and conf > 0.5:
            helmet_detected = True
            annotate_image(annotated_image, (x1, y1, x2, y2), "Helmet", (0, 255, 0))

        elif class_name in ["numberplate", "number_plate"] and conf > 0.4:
            annotate_image(annotated_image, (x1, y1, x2, y2), "Plate", (255, 0, 0))
            cropped_plate = image[y1:y2, x1:x2]
            processed_plate = preprocess_plate(cropped_plate)

            # Debug images
            cv2.imwrite("debug_crop.jpg", cropped_plate)
            cv2.imwrite("debug_processed.jpg", processed_plate)

            ocr_result = ocr.ocr(processed_plate, cls=True)

            # Updated OCR handling and plate building
            if ocr_result and isinstance(ocr_result, list) and len(ocr_result) > 0:
                full_text_parts = []
                for line in ocr_result[0]:
                    raw_text = line[1][0].upper()
                    confidence = line[1][1]
                    cleaned_text = re.sub(r'[^A-Z0-9]', '', raw_text)
                    print(f"OCR: '{raw_text}' â†’ '{cleaned_text}' (conf: {confidence:.2f})")
                    full_text_parts.append(cleaned_text)

                full_plate = ''.join(full_text_parts)
                print(f"\nâœ… Final Detected Plate Number: {full_plate}")
                if plate_pattern.match(full_plate):
                    number_plate_texts.append(full_plate)

    # Save annotated image
    annotated_path = os.path.join("annotated_image.jpg")
    cv2.imwrite(annotated_path, annotated_image)

    # Print summary
    print("\n--- Summary ---")
    print(f"Helmet Detected: {'âœ… Yes' if helmet_detected else 'âŒ No'}")
    if number_plate_texts:
        print(f"Number Plate(s): âœ… {', '.join(number_plate_texts)}")
    else:
        print("Number Plate(s): âŒ None detected")
    print(f"Annotated image saved to: {annotated_path}")

# ---------------- Run ----------------
process_image(IMAGE_PATH)
```

```python
import cv2
import re
import os
import numpy as np
from paddleocr import PaddleOCR
from ultralytics import YOLO

# ---------------- Configuration ----------------
YOLO_MODEL_PATH = "/content/runs/detect/train/weights/best.pt"  # Your YOLO model path
IMAGE_PATH = "/veihcle.jpg"  # Your test image path

# Indian number plate regex pattern (basic)
plate_pattern = re.compile(r'^[A-Z]{2}\d{1,2}[A-Z]{1,2}\d{1,4}$')

# ---------------- Model Initialization ----------------
# Load YOLOv8 model
model = YOLO(YOLO_MODEL_PATH)

# Initialize PaddleOCR
ocr = PaddleOCR(use_angle_cls=True, lang='en', show_log=False)

# ---------------- Preprocessing Function ----------------
def preprocess_plate(plate_img):
    # Step 1: Resize
    plate_img = cv2.resize(plate_img, (0, 0), fx=2.5, fy=2.5, interpolation=cv2.INTER_CUBIC)

    # Step 2: Grayscale
    gray = cv2.cvtColor(plate_img, cv2.COLOR_BGR2GRAY)

    # Step 3: Denoise
    denoised = cv2.bilateralFilter(gray, 11, 17, 17)

    # Step 4: CLAHE
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    enhanced = clahe.apply(denoised)

    # Step 5: Sharpen
    kernel_sharp = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
    sharpened = cv2.filter2D(enhanced, -1, kernel_sharp)

    # Step 6: Threshold
    _, thresh = cv2.threshold(sharpened, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Step 7: Morph Closing
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
    morph = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

    # Step 8: Padding
    padded = cv2.copyMakeBorder(morph, 15, 15, 15, 15, cv2.BORDER_CONSTANT, value=255)

    return padded

# ---------------- Annotation Helper ----------------
def annotate_image(img, box, label, color=(0, 255, 0)):
    x1, y1, x2, y2 = box
    cv2.rectangle(img, (x1, y1), (x2, y2), color, 2)
    cv2.putText(img, label, (x1, y1 - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

# ---------------- Main Pipeline ----------------
def process_image(image_path):
    image = cv2.imread(image_path)
    annotated_image = image.copy()
    number_plate_texts = []
    helmet_detected = False

    results = model(image_path)[0]

    for box in results.boxes:
        cls_id = int(box.cls[0])
        conf = float(box.conf[0])
        class_name = results.names[cls_id].lower()
        x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())

        if class_name == "helmet" and conf > 0.5:
            helmet_detected = True
            annotate_image(annotated_image, (x1, y1, x2, y2), "Helmet", (0, 255, 0))

        elif class_name in ["numberplate", "number_plate"] and conf > 0.4:
            annotate_image(annotated_image, (x1, y1, x2, y2), "Plate", (255, 0, 0))
            cropped_plate = image[y1:y2, x1:x2]
            processed_plate = preprocess_plate(cropped_plate)

            # Optional debug
            cv2.imwrite("debug_crop.jpg", cropped_plate)
            cv2.imwrite("debug_processed.jpg", processed_plate)

            ocr_result = ocr.ocr(processed_plate, cls=True)

            # OCR result cleaning and joining
            full_plate = ""
            if ocr_result and isinstance(ocr_result, list) and len(ocr_result) > 0:
                for line in ocr_result[0]:
                    raw_text = line[1][0].upper()
                    confidence = line[1][1]
                    cleaned_text = re.sub(r'[^A-Z0-9]', '', raw_text)

                    if len(cleaned_text) < 2 or confidence < 0.6:
                        continue  # Skip noise
                    print(f"OCR: '{raw_text}' â†’ '{cleaned_text}' (conf: {confidence:.2f})")
                    full_plate += cleaned_text

                full_plate = re.sub(r'[^A-Z0-9]', '', full_plate)

                if plate_pattern.match(full_plate):
                    number_plate_texts.append(full_plate)
                    print(f"\nâœ… Final Detected Plate Number: {full_plate}")
                elif 8 <= len(full_plate) <= 13 and full_plate.isalnum():
                    number_plate_texts.append(full_plate)
                    print(f"\nâš ï¸ Final Detected Plate (non-matching): {full_plate}")

    # Save output
    annotated_path = os.path.join("annotated_image.jpg")
    cv2.imwrite(annotated_path, annotated_image)

    # Summary
    print("\n--- Summary ---")
    print(f"Helmet Detected: {'âœ… Yes' if helmet_detected else 'âŒ No'}")
    if number_plate_texts:
        print(f"Number Plate(s): âœ… {', '.join(number_plate_texts)}")
    else:
        print("Number Plate(s): âŒ None detected")
    print(f"Annotated image saved to: {annotated_path}")

# ---------------- Run ----------------
process_image(IMAGE_PATH)
```
