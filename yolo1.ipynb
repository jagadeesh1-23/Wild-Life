```python
!mkdir -p datasets && cd datasets
!wget https://github.com/ultralytics/assets/releases/download/v0.0.0/african-wildlife.zip
!unzip african-wildlife.zip -d african-wildlife
```

```python
# Content of the data.yaml file
data_yaml = """
path: ../datasets/african-wildlife
train: /content/african-wildlife/train/images
val: /content/african-wildlife/valid/images
test: /content/african-wildlife/test/images

names:
  0: buffalo
  1: elephant
  2: rhino
  3: zebra
"""

# Save the file to the working directory
with open("african-wildlife.yaml", "w") as f:
    f.write(data_yaml)

print("data.yaml created successfully!")
```
